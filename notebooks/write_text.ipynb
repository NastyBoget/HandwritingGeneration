```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((1000, 1000, 3), dtype=np.uint8)

ft = cv2.freetype.createFreeType2()
ft.loadFontData(fontFileName='../fonts/0000.ttf',
                id=0)
ft.putText(img=img,
           text='    ,',
           org=(15, 170),
           fontHeight=100,
           color=(255, 255, 255),
           thickness=-1,
           line_type=cv2.LINE_AA,
           bottomLeftOrigin=True)
ft.putText(img=img,
           text='    ',
           org=(15, 270),
           fontHeight=100,
           color=(255, 255, 255),
           thickness=-1,
           line_type=cv2.LINE_AA,
           bottomLeftOrigin=True)
ft.putText(img=img,
           text=' ,  ',
           org=(15, 370),
           fontHeight=100,
           color=(255, 255, 255),
           thickness=-1,
           line_type=cv2.LINE_AA,
           bottomLeftOrigin=True)
ft.putText(img=img,
           text=' ,  ,  .',
           org=(15, 470),
           fontHeight=100,
           color=(255, 255, 255),
           thickness=-1,
           line_type=cv2.LINE_AA,
           bottomLeftOrigin=True)

plt.figure(figsize=(10, 10))
plt.imshow(img)
```
