```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((1000, 1000, 3), dtype=np.uint8)

ft = cv2.freetype.createFreeType2()
ft.loadFontData(fontFileName='../fonts/0000.ttf',
                id=0)
ft.putText(img=img,
           text='И сердце бьется в упоенье,',
           org=(15, 170),
           fontHeight=100,
           color=(255, 255, 255),
           thickness=-1,
           line_type=cv2.LINE_AA,
           bottomLeftOrigin=True)
ft.putText(img=img,
           text='И для него воскресли вновь',
           org=(15, 270),
           fontHeight=100,
           color=(255, 255, 255),
           thickness=-1,
           line_type=cv2.LINE_AA,
           bottomLeftOrigin=True)
ft.putText(img=img,
           text='И божество, и вдохновенье',
           org=(15, 370),
           fontHeight=100,
           color=(255, 255, 255),
           thickness=-1,
           line_type=cv2.LINE_AA,
           bottomLeftOrigin=True)
ft.putText(img=img,
           text='И жизнь, и слезы, и любовь.',
           org=(15, 470),
           fontHeight=100,
           color=(255, 255, 255),
           thickness=-1,
           line_type=cv2.LINE_AA,
           bottomLeftOrigin=True)

plt.figure(figsize=(10, 10))
plt.imshow(img)
```

```python

img = np.zeros((1000, 1000, 3), dtype=np.uint8)

for j in range(5):
    for i in range(5):
        ft.putText(img=img, text='а', org=(15 + i * 100, 150 + j * 100),
               fontHeight=100, color=(255, 255, 255), thickness=-1,
               line_type=cv2.LINE_AA, bottomLeftOrigin=True)

plt.imshow(img)
```

```python
from PIL import Image, ImageFont, ImageDraw

img = Image.fromarray(np.zeros((1000, 1000, 3), dtype=np.uint8))
draw = ImageDraw.Draw(img)

font = ImageFont.truetype("../fonts/0000.ttf", 130)
print(draw.textbbox((0, 0), "И сердце бьется в упоенье,", font))
draw.text((0, 0), "И сердце бьется в упоенье,", (255, 255, 255), font=font)

draw.text((0, 150), "И для него воскресли вновь", (255, 255, 255), font=font)

draw.text((0, 300), "И божество, и вдохновенье", (255, 255, 255), font=font)

plt.imshow(img)
```
